```python
import math

# Y-axis values
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2009)

# getting the mean
m = ys / len(s2009)

# looping thru the array 
for i in s2009:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2009) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2009 = {SD:.2f}')



### 2010 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2010)

# getting the mean
m = ys / len(s2010)

# looping thru the array 
for i in s2010:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2010) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2010 = {SD:.2f}')


### S2011 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2011)

# getting the mean
m = ys / len(s2011)

# looping thru the array 
for i in s2011:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2011) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2011 = {SD:.2f}')

### 2012 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2012)

# getting the mean
m = ys / len(s2012)

# looping thru the array 
for i in s2012:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2012) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2012 = {SD:.2f}')

### 2013 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2013)

# getting the mean
m = ys / len(s2013)

# looping thru the array 
for i in s2013:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2013) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2013= {SD:.2f}')

### 2014 YEAR ###
sqs =[]
 #standard deviation

# adding all the number in the array
ys= sum(s2014)

# getting the mean
m = ys / len(s2014)

# looping thru the array 
for i in s2014:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2014) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2014 = {SD:.2f}')

### 2015 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2015)

# getting the mean
m = ys / len(s2015)

# looping thru the array 
for i in s2015:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2015) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2015 = {SD:.2f}')

### 2016 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2016)

# getting the mean
m = ys / len(s2016)

# looping thru the array 
for i in s2016:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2016) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2016 = {SD:.2f}')

### 2017 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2017)

# getting the mean
m = ys / len(s2017)

# looping thru the array 
for i in s2017:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2017) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2017= {SD:.2f}')

### 2018 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2018)

# getting the mean
m = ys / len(s2018)

# looping thru the array 
for i in s2018:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2018) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2018 = {SD:.2f}')

### 2019 YEAR ###
sqs =[]
# standard deviation

# adding all the number in the array
ys= sum(s2019)

# getting the mean
m = ys / len(s2019)

# looping thru the array 
for i in s2019:
    # adding each number in the array with the mean 
    sq = (i - m) ** 2
    sqs.append(sq)

# summation of the numerator
n = sum(sqs)

# 
z = n/ (len(s2019) - 1) 

# Standard deviation 
SD = math.sqrt(z)
print(f'standard deviation 2019 = {SD:.2f}')
```
